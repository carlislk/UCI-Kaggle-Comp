```python
# import csv

# with open("rainfall/data/s_te_err.csv") as f:
#     reader = csv.DictReader(f)
#     rows = [row for row in reader ]

# for row in rows:
#     print (row)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

f_df = pd.read_csv("rainfall/data/f_tr_err.csv")
tr_df = pd.read_csv("rainfall/data/s_tr_err.csv")
te_df = pd.read_csv("rainfall/data/s_te_err.csv")

# Select Columns
# MAE
mae_f = f_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mae']]
mae_tr = tr_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mae']]
mae_te = te_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mae']]
print(mae_f.columns)
print(mae_tr.columns)
print(mae_te.columns)



# ------------------------------------------------------------------------------------
# Depth - Estimator = 900, Alpha = 1
depth_mae_a_f = mae_f[(mae_f['Estimator'] == 900) & (mae_f['Alpha'] == 1)].loc[:,[ 'Depth', 'mae']]
depth_mae_a_tr = mae_tr[(mae_tr['Estimator'] == 900) & (mae_tr['Alpha'] == 1)].loc[:,[ 'Depth', 'mae']]
depth_mae_a_te = mae_te[(mae_te['Estimator'] == 900) & (mae_te['Alpha'] == 1)].loc[:,[ 'Depth', 'mae']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MAE on Depth - Est=900, Alpha=1")
plt.xlabel("Depth")
plt.ylabel("MAE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mae_depth.png")
plt.clf()


# ------------------------------------------------------------------------------------
# Estimator - Depth 12 Alpha = 1
depth_mae_a_f = mae_f[(mae_f['Depth'] == 12) & (mae_f['Alpha'] == 1)].loc[:,[ 'Estimator', 'mae']]
depth_mae_a_tr = mae_tr[(mae_tr['Depth'] == 12) & (mae_tr['Alpha'] == 1)].loc[:,[ 'Estimator', 'mae']]
depth_mae_a_te = mae_te[(mae_te['Depth'] == 12) & (mae_te['Alpha'] == 1)].loc[:,[ 'Estimator', 'mae']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MAE on Estimator - Depth=12, Alpha=1")
plt.xlabel("Estimator")
plt.ylabel("MAE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mae_est.png")
plt.clf()


# ------------------------------------------------------------------------------------
# Alpha - Est 900, Depth 12
depth_mae_a_f = mae_f[(mae_f['Depth'] == 12) & (mae_f['Estimator'] == 900)].loc[:,[ 'Alpha', 'mae']]
depth_mae_a_tr = mae_tr[(mae_tr['Depth'] == 12) & (mae_tr['Estimator'] == 900)].loc[:,[ 'Alpha', 'mae']]
depth_mae_a_te = mae_te[(mae_te['Depth'] == 12) & (mae_te['Estimator'] == 900)].loc[:,[ 'Alpha', 'mae']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MAE on Alpha - Depth=12, Est=900")
plt.xlabel("Alpha")
plt.ylabel("MAE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mae_alpha.png")
plt.clf()











# MSE
mse_f = f_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mse']]
mse_tr = tr_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mse']]
mse_te = te_df.loc[:,['Depth', 'Estimator', 'Alpha', 'mse']]
print(mse_f.columns)
print(mse_tr.columns)
print(mse_te.columns)



# ------------------------------------------------------------------------------------
# Depth - Estimator = 900, Alpha = 1
depth_mae_a_f = mse_f[(mse_f['Estimator'] == 900) & (mse_f['Alpha'] == 1)].loc[:,[ 'Depth', 'mse']]
depth_mae_a_tr = mse_tr[(mse_tr['Estimator'] == 900) & (mse_tr['Alpha'] == 1)].loc[:,[ 'Depth', 'mse']]
depth_mae_a_te = mse_te[(mse_te['Estimator'] == 900) & (mse_te['Alpha'] == 1)].loc[:,[ 'Depth', 'mse']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MSE on Depth - Est=900, Alpha=1")
plt.xlabel("Depth")
plt.ylabel("MSE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mse_depth.png")
plt.clf()


# ------------------------------------------------------------------------------------
# Estimator - Depth 12 Alpha = 1
depth_mae_a_f = mse_f[(mse_f['Depth'] == 12) & (mse_f['Alpha'] == 1)].loc[:,[ 'Estimator', 'mse']]
depth_mae_a_tr = mse_tr[(mse_tr['Depth'] == 12) & (mse_tr['Alpha'] == 1)].loc[:,[ 'Estimator', 'mse']]
depth_mae_a_te = mse_te[(mse_te['Depth'] == 12) & (mse_te['Alpha'] == 1)].loc[:,[ 'Estimator', 'mse']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MSE on Estimator - Depth=12, Alpha=1")
plt.xlabel("Estimator")
plt.ylabel("MSE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mse_est.png")
plt.clf()


# ------------------------------------------------------------------------------------
# Alpha - Est 900, Depth 12
depth_mae_a_f = mse_f[(mse_f['Depth'] == 12) & (mse_f['Estimator'] == 900)].loc[:,[ 'Alpha', 'mse']]
depth_mae_a_tr = mse_tr[(mse_tr['Depth'] == 12) & (mse_tr['Estimator'] == 900)].loc[:,[ 'Alpha', 'mse']]
depth_mae_a_te = mse_te[(mse_te['Depth'] == 12) & (mse_te['Estimator'] == 900)].loc[:,[ 'Alpha', 'mse']]

# Convert To Numpy
np_mae_f = np.array(depth_mae_a_f)
np_mae_tr = np.array(depth_mae_a_tr)
np_mae_te = np.array(depth_mae_a_te)


# Plot
plt.title("MSE on Alpha - Depth=12, Est=900")
plt.xlabel("Alpha")
plt.ylabel("MSE")
plt.semilogy(np_mae_f[:,0],np_mae_f[:,1], color='green', label="Full Data")
plt.semilogy(np_mae_tr[:,0],np_mae_tr[:,1], color='blue', label="Train")
plt.semilogy(np_mae_te[:,0],np_mae_te[:,1], color='red', label="Test")
plt.legend(loc=0)
#plt.show()
plt.savefig("graphs_1/mse_alpha.png")
plt.clf()


```
